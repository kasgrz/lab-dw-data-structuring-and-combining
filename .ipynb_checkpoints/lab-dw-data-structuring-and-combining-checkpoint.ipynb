Observation:

- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import numpy as np
import pandas as pd

file1 = pd.read_csv("cleaned_data.csv")

file2 = pd.read_csv(
    "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
)

file3 = pd.read_csv(
    "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
)
```

```python
file1.head()
```

```python
file2.head()
```

```python
file3.head()
```

```python
def clean_data(df):
    df.columns = df.columns.str.lower().str.replace(" ", "_") 
    df = df.rename(columns={"st": "state"})
    df["gender"] = df["gender"].replace({"Male":"M", "female":"F", "Femal":"F"
})
    df["state"] = df["state"].replace({"AZ":"Arizona", "WA":"Washington", "Cali":"California"
})
    df["education"] = df["education"].replace({"Bachelors":"Bachelor"
})
    df["vehicle_class"] = df["vehicle_class"].replace({"Sports Car":"Luxury", "Luxury SUV":"Luxury", "Luxury Car":"Luxury"
})
    df["customer_lifetime_value"] = (
    df["customer_lifetime_value"]
    .astype(str)
    .str.replace("%", "", regex=False)
)
    
    df["customer_lifetime_value"] = df["customer_lifetime_value"].str.replace("%", "", regex=False)
    df = df.drop_duplicates()

    df["customer_lifetime_value"] = pd.to_numeric(
    df["customer_lifetime_value"],
    errors="coerce"
)
    df["gender"] = df["gender"].fillna("Unknown")
    df = df.dropna(how="all")
    
    return df

file2 = clean_data(file2)
file3 = clean_data(file3)

```

```python
file3.head()
```

```python
file3["number_of_open_complaints"].unique()
```

```python
file2["number_of_open_complaints"] = file2["number_of_open_complaints"].str.split("/").str[1]
```

```python
file2["number_of_open_complaints"].unique()
```

```python
file2["number_of_open_complaints"]=file2["number_of_open_complaints"].astype(int)
```

```python
combined = pd.concat([file1, file2, file3], ignore_index=True)
```

```python
combined.shape
```

```python
combined = combined.drop_duplicates()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
df.columns
```

```python
df.head()
```

```python
revenue_by_channel = pd.pivot_table(
    df,
    index="sales_channel",
    values="total_claim_amount",
    aggfunc="sum"
).round(2)

revenue_by_channel
```

```python
# Agent generated the highest total revenue, contributing approximately 38% of the total sales revenue. This suggests that direct interaction with customers through agents remains the most effective sales channel.
# Branch offices generated the second-highest revenue, indicating that traditional in-person sales channels continue to play an important role.
# The Call Center generated moderate revenue, outperforming the Web channel, which suggests that customers may prefer human interaction when purchasing insurance products.
# The Web channel produced the lowest total revenue. This could indicate lower customer adoption of digital channels or potential opportunities to improve online marketing and customer experience.
```

```python
df = df.drop_duplicates()
df = df.dropna(how="all")
df.info()
```

```python
clv_pivot = pd.pivot_table(
    df,
    index="gender",
    columns="education",
    values="customer_lifetime_value",
    aggfunc="mean"
).round(2)

clv_pivot
```

```python
# Customers with "High School or Below" education have the highest average CLV, particularly among female customers (8675.22). This suggests that lower educational attainment does not necessarily correspond to lower customer value.
# Customers with doctoral degrees have the lowest average CLV for both genders, indicating that highly educated customers are not the most profitable segment in this dataset.
# Gender differences are relatively small across most education levels. For example, male and female customers with master's degrees have nearly identical CLV values.
# The relationship between education and customer lifetime value is not linear. Higher educational attainment does not consistently lead to higher customer value.
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
complaints = (
    df.groupby(["policy_type", "month"])
      .size()
      .reset_index(name="number_of_complaints")
)

complaints
```

```python
#Personal Auto policies generated the highest number of complaints in both months, accounting for the majority of customer complaints. This is likely due to Personal Auto being the company's largest product category.
#Corporate Auto policies generated a substantially lower number of complaints than Personal Auto but still represented a significant share of customer service interactions.
#Special Auto policies generated the fewest complaints, suggesting either a smaller customer base or fewer service-related issues.
#For all policy types, the number of complaints decreased from month 1 to month 2:
#Personal Auto complaints decreased from 4,329 to 3,799.
#Corporate Auto complaints decreased from 1,252 to 1,089.
#Special Auto complaints decreased from 237 to 204.
```
